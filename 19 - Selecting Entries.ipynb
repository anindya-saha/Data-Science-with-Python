```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

#### Selecting entries from a Series

```python
ser1 = Series(np.arange(3), index=['A', 'B', 'C'])

# double every value in the series
ser1 = 2 * ser1

ser1
```

```python
# grab value by index name
ser1['B']
```

```python
# grab value by index ordinal
ser1[1]
```

```python
# grab value by index range
ser1[0:3]
```

```python
# grab value by range of index values
ser1[['A', 'B']]
```

```python
# grab value by boolean value of logic, get values where its greater than 3
ser1[ser1 > 3]
```

```python
# overwriting entries in the series
ser1[ser1 > 3] = 10

ser1
```

#### Selecting entries from a Dataframe

```python
dframe1 = DataFrame(np.arange(25).reshape(5,5), index=['NYC', 'LA', 'SF','DC', 'CHI'], 
                    columns=['A', 'B', 'C', 'D', 'E'])

dframe1
```

```python
# select data by a column name
dframe1['B']
```

```python
# select data by list of column names
dframe1[['B','E']]
```

```python
# grab value by boolean value of logic, get values from column C where its greater than 8
dframe1[dframe1['C'] > 8]
```

```python
# we can also show a boolean dataframe, it will just have boolean values true and false
dframe1 > 10
```
