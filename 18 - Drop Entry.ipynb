```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

#### Dropping from a Series

```python
ser1 = Series(np.arange(3), index=['a', 'b', 'c'])
                                      
ser1
```

```python
# drop the index 'b'
ser1.drop('b')
```

#### Dropping from a DataFrame

```python
dframe1 = DataFrame(np.arange(9).reshape(3,3), index=['SF', 'LA', 'NY'], columns=['population', 'size', 'year'])

dframe1
```

```python
# drop a row with index 'SF'
dframe1.drop('SF')
```

```python
# drop does not affect the original dframe
dframe1
```

```python
dframe2 = dframe1.drop('SF')

dframe2
```

```python
# incase we want to drop a column we need to mention axis = 1. By default axis = 0 which means row
dframe1.drop('year', axis=1)
```
